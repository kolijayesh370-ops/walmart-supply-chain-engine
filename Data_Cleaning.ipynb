```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Walmart_Sales.csv")
```

```python
print("Shape of the dataset:", df.shape)

print("\nColumns:", df.columns.tolist())

df.head()
```

```python
df.info()

df.dtypes
```

```python
df['Store'] = df['Store'].astype(object)

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

df.dtypes
```

```python
print("Missing values count before handling:")
print(df.isna().sum())

df['Weekly_Sales'] = df['Weekly_Sales'].fillna(df['Weekly_Sales'].median())
df['Temperature'] = df['Temperature'].fillna(df['Temperature'].median())
df['Fuel_Price'] = df['Fuel_Price'].fillna(df['Fuel_Price'].median())

df.dropna(inplace=True)

print("\nMissing values count after cleaning:")
print(df.isna().sum())
```

```python
print("Duplicate count:", df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("Duplicate count after removal:", df.duplicated().sum())
```

```python
Quality_report = pd.DataFrame({
    'column': df.columns,
    'Data Types': df.dtypes.values,
    'Missing Values': df.isna().sum().values,
    'Unique Values': df.nunique().values
})
Quality_report
```

```python
df.to_csv("Walmart_Sales_Cleaned.csv", index=False)
print("Cleaned Dataset created successfully!")
```
